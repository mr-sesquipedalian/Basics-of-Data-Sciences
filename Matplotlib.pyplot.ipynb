```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
x=np.linspace(-2*math.pi,2*math.pi)
y=np.sin(x)
z=np.cos(x)
plt.plot(x,y,color="red")
plt.plot(x,z)
plt.xlabel("value")
plt.ylabel("angle(in radians)")
plt.title("ratios chart")
plt.grid()
plt.axhline()
plt.axvline()
plt.show()
```

### Histogram

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate a normal distribution (1-D)
u=5
sigma=1
#Such distrbution is called standard normal distribution
vals=u+sigma*np.random.randn(1000)   #argument passed is the number of elements in sample space

plt.hist(vals,10)       #argument passed in numerical form is the number of class interval available
plt.show()
```

```python
vals=np.round(vals)
z=np.unique(vals, return_counts=True)
print(z)
```

### Multivariate normal distribution

```python
#bivariate
mean=np.array([0.0,0.0])
cov=np.array([[1,0],[0,1]])
dist=np.random.multivariate_normal(mean,cov,500)
print(dist.shape)
plt.scatter (dist[:,0],dist[:,1])
```

```python
plt.scatter (dist[:,0],dist[:,1])
```

```python
#bivariate
mean=np.array([0.0,0.0])
cov=np.array([[1,0.8],[0.8,1]])

mean2=np.array([5.0,6.0])
cov2=np.array([[1.3,0.2],[0.2,1.1]])

dist=np.random.multivariate_normal(mean,cov,500)
dist2=np.random.multivariate_normal(mean2,cov2,500)
print(dist.shape)
plt.scatter (dist[:,0],dist[:,1])
plt.scatter(dist2[:,0],dist2[:,1])
plt.show()
```
