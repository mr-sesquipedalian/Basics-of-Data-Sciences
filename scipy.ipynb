```python
import matplotlib.pyplot as plt
```

```python
r=np.random.randn(10000)
plt.hist(r,bins=100,histtype="bar",ec="black")
plt.show()
```

```python
# IF THE DATA IS NOT STANDARD NORMAL BUT IT HAS SOME FINITE MEAN AND STANDARD DEVIATION
R=10*np.random.randn(10000)+5
plt.hist(R, bins=100,histtype="bar",ec="black")
plt.show()
```

```python
f=np.random.randn(10000,2)
```

```python
plt.scatter(f[:,0],f[:,1])#since our array is 2-D now we will consider both the axis as separate and wil designate them accordingly
plt.show()
```

```python
cov=np.array([[1,0.8],[0.8,3]])
from scipy.stats import multivariate_normal as mvn
```

```python
mu=np.array([0,2])
```

```python
r=mvn.rvs(mean=mu, cov=cov, size=1000)
```

```python
plt.axis('equal')
plt.scatter(r[:,0],r[:,1])
```
